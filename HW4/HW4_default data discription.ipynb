```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터셋 로드
data = pd.read_csv("Default.csv")

# Unnamed: 0 열 제거
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]

# 1. 데이터셋의 기본 정보 출력 (변수명, 데이터 타입, Null 값 개수 등)
print("1. 데이터 정보:")
data.info()


# 2. 데이터셋의 기술 통계량 출력 (수치형 변수의 평균, 표준편차, 최소/최대값 등 확인)
print("\n3. 기술 통계량:")
print(data.describe(include="all"))

# 3. 각 변수의 고유값 개수 파악
print("\n4. 고유값 개수:")
for column in data.columns:
    unique_values = data[column].nunique()
    print(f"{column}: {unique_values} unique values")

# 4. 범주형 변수 분포 확인 (default와 student 변수)
print("\n5. Categorical Variable Distribution:")
print("default 분포:\n", data["default"].value_counts())
print("\nstudent 분포:\n", data["student"].value_counts())

# 5. 결측값 확인
print("\n6. 결측값 개수:")
print(data.isnull().sum())

# 6. 각 수치형 변수의 히스토그램을 통해 분포 시각화
numeric_columns = data.select_dtypes(include=["float64", "int64"]).columns
data[numeric_columns].hist(bins=30, figsize=(10, 8))
plt.suptitle("Numerical Variables Distribution")
plt.show()
```
