```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터셋 로드
data = pd.read_csv("Weekly.csv")

# 1. 데이터셋의 기본 정보 출력 (변수명, 데이터 타입, Null 값 개수 등)
print("1. 데이터 정보:")
data.info()

# 3. 데이터셋의 기술 통계량 출력 (수치형 변수의 평균, 표준편차, 최소/최대값 등 확인)
print("\n2. 기술 통계량:")
print(data.describe(include="all"))

# 4. 각 변수의 고유값 개수 파악
print("\n3. 고유값 개수:")
for column in data.columns:
    unique_values = data[column].nunique()
    print(f"{column}: {unique_values} unique values")

# 5. 범주형 변수 분포 확인 (Direction 변수)
print("\n4. Categorical Variable Distribution:")
print("Direction 분포:\n", data["Direction"].value_counts())

# 6. 결측값 확인
print("\n5. 결측값 개수:")
print(data.isnull().sum())

# 7. 수치형 변수의 히스토그램을 통한 분포 시각화
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
data[numeric_columns].hist(bins=30, figsize=(15, 10))
plt.suptitle("Numerical Variables Distribution")
plt.show()

# 8. Direction 변수의 분포 시각화 (막대그래프)
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Direction')
plt.title('Direction 변수의 분포')
plt.xlabel('Direction')
plt.ylabel('Count')
plt.show()

# 9. 상관관계 히트맵
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```
