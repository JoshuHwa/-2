```python
# Step 1: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the dataset (if not already loaded)
# Replace 'file_path' with your actual file path if running separately
file_path = 'C:/Users/bikmi/Desktop/Bigdata HW/Boston.csv'  # Use the correct path for your dataset
df = pd.read_csv(file_path)

# Step 3: Filter for quantitative variables
quantitative_cols = df.select_dtypes(include=['float64', 'int64']).columns
quantitative_df = df[quantitative_cols]

# Step 4: Check the correlation matrix of the quantitative variables
correlation_matrix = quantitative_df.corr()

# Step 5: Visualize the correlation matrix with a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Quantitative Variables')
plt.show()

# Step 6: Produce scatter plots for some of the highest correlated variables
# Example: Scatter plot for variables with high correlation
# You can choose highly correlated pairs from the heatmap
sns.pairplot(quantitative_df[['MEDV', 'RM', 'LSTAT', 'PTRATIO', 'CRIM']])
plt.suptitle('Scatter Plot of Selected Quantitative Variables')
plt.show()
```

```python
# Step 1: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the dataset (if not already loaded)
file_path = 'C:/Users/bikmi/Desktop/Bigdata HW/Boston.csv'  # Use the correct path for your dataset
df = pd.read_csv(file_path)

# Step 3: Filter for quantitative variables
quantitative_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Step 4: Provide a graph for quantitative variables (histograms)
fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(16, 12))  # 5 rows and 4 columns
axes = axes.flatten()  # Flatten the axes to easily iterate over them

# Plot each quantitative variable in its corresponding subplot
for i, col in enumerate(quantitative_cols):
    df[col].hist(bins=15, edgecolor='black', ax=axes[i])
    axes[i].set_title(col)

# Remove any unused axes (because we have 19 plots but defined space for 20)
for j in range(len(quantitative_cols), len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('Histograms of Quantitative Variables')
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to prevent overlap
plt.show()
```
