Discriptive Stastics of Auto Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

auto = pd.read_csv("C:/Users/bikmi/Desktop/Bigdata_HW/HW2/Auto.csv")


# '?' 값을 NaN으로 변환
auto["horsepower"] = auto["horsepower"].replace("?", np.nan)

# NaN 처리: 필요에 따라 삭제하거나 평균값으로 채움
auto["horsepower"] = auto["horsepower"].astype(float)

# 필요시 NaN 값을 평균으로 대체
auto["horsepower"].fillna(auto["horsepower"].mean(), inplace=True)

print(auto.info())
```

```python
# 기본 통계량 확인
print("\n기초 통계량:")
print(auto.describe())
```

```python
# 히스토그램
auto.hist(bins=15, figsize=(15, 10))
plt.suptitle("Distribution of variables in Auto Data", fontsize=16)
plt.show()
```
